```python
import nltk

file = open ("cacm.txt","r")
#count = 1
for line in file:
    #print(line)
    wordlist = list(nltk.bigrams(line.split())) #bigram
    #wordlist = line.split() #single word
    #items = line.split() #all possible pairs
    #wordlist = [(items[i],items[j]) for i in range(len(items)) for j in range(i+1, len(items))] #all possible pairs
    
    wordfreq = []
    for w in wordlist:
        wordfreq.append(wordlist.count(w))
    print(str(list(zip(wordlist, wordfreq))))
    
    #To separate documents
    print("\n")
#print(count)
```
