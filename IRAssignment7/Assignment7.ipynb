```python
import os.path
import matplotlib.pyplot as plt
```

```python
#Preprocessing
class DataRead(object):
    def preprocess(self):
        path = './Fold1/'
        filepath = os.path.join(path,"train.txt")
        wpath = './Data/'
        with open(filepath) as fp:
            trainFile = os.path.join(wpath,"train.txt")
            trainFile = open(trainFile,"w")
            
            validFile = os.path.join(wpath,"valid.txt")
            validFile = open(validFile,"w")
            
            testFile = os.path.join(wpath,"test.txt")
            testFile = open(testFile,"w")
            
            for i in range(0, 1001*50):
                if (i<600*50):
                    line = fp.readline()
                    trainFile.write(line)
                if(i>700*50 and 1< 801*50):
                    line = fp.readline()
                    validFile.write(line)
                if (i>800*50 and i< 1001*50):
                    line = fp.readline()
                    testFile.write(line)
                else:
                    fp.readline()
                    
```

```python
#main function
if __name__ == "__main__":
    DataRead().preprocess()
```

```python
x = []
with open('./output/RankBoost.ndcg.txt', 'r') as file:
    for row in file:
        a, b, c = row.split()
        x.append(float(c))
plt.hist(x, bins=10)
plt.suptitle('RankBoost')
plt.xlabel('NDCG@10')
plt.ylabel('Frequency')
plt.show()
```
